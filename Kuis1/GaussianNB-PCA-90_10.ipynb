## Import Library

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA as RandomizedPCA
from sklearn.pipeline import make_pipeline
```

## Sample Data

Melihat sample data pada mnist serta memvisualisasikan data tersebut

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()

# Mengunduh dataset MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=False)

# Mengambil data dan label
X = mnist.data
y = mnist.target

# Menampilkan 15 citra pertama berserta labelnya
plt.figure(figsize=(12, 5))
for i in range(15):
    plt.subplot(3, 5, i + 1)
    plt.imshow(X[i].reshape(28, 28), cmap='gray')
    plt.title(f'Label: {y[i]}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Pembagian Data Train dan Data Test

Untuk melatih model yang akan dibuat, maka diperlukan data train dan data test terlebih dahulu. Kami menggunakan komposisi 10% untuk data test dan 90% untuk data train

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.1,
                                                random_state=42)
```

Setalah membagi data tersebut, maka dapat dilakukan proses training. Kami menggunakan **pca** untuk meminimalisir fitur yang penting yang akan digunakan. Lalu kami menggunakan GaussianNB untuk metode dalam pembuatan model. Kemudian untuk menggabungkan kedua proses tersebut dan melakukan proses iteratif untuk membuat model, mendapatkan hasil, dan meningkatkan kualitas model, maka diperlukan fungsi make_pipeline(). Lalu kita berikan semua data train kita pada model dengan menggunakan model.fit()

```python
pca = RandomizedPCA(n_components=150, whiten=True, random_state=42)
nb = GaussianNB()
model = make_pipeline(pca, nb)
model.fit(Xtrain, ytrain)
```

## Melakukan prediksi pada test set

membentuk model, maka selanjutnya dalah melakukan prediksi dengan datatest yang dimiliki

```python
y_pred = model.predict(Xtest)
print("Number of mislabeled points out of a total %d points : %d"% (Xtest.shape[0], (ytest != y_pred).sum()))
```

## Melihat Akurasi Model

```python
print(classification_report(ytest, y_pred))
```

Dari `classification_report` didapat akurasi model dengan menggunakan metode tersebut adalah **85%**

## Confusion Matrix
`confusion_matrix()` bertujuan untuk melihat berapa data yang sesuai antara data prediksi sama data test

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
mat = confusion_matrix(ytest, y_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=mnist.target_names,
            yticklabels=mnist.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label')
```

Hasilnya, mayoritas data test mampu diprediksi oleh model dengan benar

## Visualisasi Data

```python
n = 30

plt.figure(figsize=(12, 6))
for i in range(n):
    plt.subplot(6, 5, i + 1)
    plt.imshow(Xtest[i].reshape(28, 28), cmap="gray")

    # Determine text color based on true or predicted
    true_label = ytest.iloc[i] if isinstance(ytest, pd.Series) else ytest[i]
    pred_label = y_pred.iloc[i] if isinstance(y_pred, pd.Series) else y_pred[i]

    if true_label == pred_label:
        text_color = 'black'
    else:
        text_color = 'red'

    plt.title(f"True: {true_label}\nPred: {pred_label}", color=text_color)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
